<h3>Tитаник през моя поглед</h3>

<b>Бърз преглед</b>
1. Премахвам колоните 'Cabin' и 'Ticket'
2. Разглеждам корелационната матрица
3. Попълвам липсващите стойности
    -  за 'Embarked' с 'S'
    -  за 'Age' със средната спрямо титлите 
4. Въвеждам нови фийчъри
    -  разделям възрастта на различни класове
    -  разделям класата на три
    -  големина на семейството и дали е пътувал сам
    -  Age\*Clas, Fare\*Class, Sex\*Class
5. Експерименти
     -  разделянето на вързастта на три класа даде най-добри резултати
     -  размера на семейство не подобри модела
     -  Age\*... не подобриха модела
     -  накрая съкратих фийчърите и резултата от модела беше малко по-добър
     -  Random Forest дава по-добри резултати от линейна, логистична регресия и SVC

Извод: Ако бях на Титаник, евентуално щях да оживея.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import mglearn

from IPython.display import display
from scipy.stats.mstats import mode

%matplotlib inline
```

```python
original = pd.read_csv('train.csv', index_col='PassengerId')
original.head(10)
```

Разглеждайки данните, виждам, че в <b>Cabin</b> колоната има най-много липсващи стойности. Тя би носила полезна информация за това в коя част на кораба са се намирали каютите, от които има най-много загинали, нo тъй като не можем да възстановим данните, ще я премахнем. Също така <b>Ticket</b>(номера на билета) смятам, че няма да ни донесе съществена информация, тъй като най-вероятно е свързан с данните за пристанище или класа, т.е. данните ще са корелирани и можем да си позволим да премахнем и нея.

```python
data = original.copy()
data = data.drop(['Cabin','Ticket'], axis=1)
```

Нека да видим корелационната матрица на непрекъснатите стойности, за да можем на пръв поглед да преценим кои фийчъри са от най-голямо значение и си струва да инвестираме време върху тях - да попълним липсващите стойности.

```python
plt.figure(figsize=(10,8))
sns.heatmap(original.corr(), vmax=0.6, square=True, annot=True)
```

Най-голямо влияние дали даден пътник ще оцелее имат: <b>Pclass</b>, <b>Age</b>, <b>SibSp</b>, <b>Parch</b>.
Ще попълним липсващите данни за възраст със средната стойност, използвайки титлите.

```python
data.head(10)
```

```python
data['Title'] = data.Name.str.extract('([A-Za-z]+)\.', expand=False)
data.loc[data.Title == 'Mlle', 'Title'] = 'Miss'
data.loc[data.Title == 'Mme', 'Title']  = 'Mrs'
data.loc[data.Title == 'Ms', 'Title']   = 'Miss'
```

```python
rare_titles = ['Dr', 'Rev', 'Col', 'Major', 'Countess', 'Don', 'Jonkheer', 'Capt', 'Lady', 'Sir']
data.Title = data.Title.replace(rare_titles, 'Rare')
```

```python
age_by_title = data.groupby('Title').Age.mean()

data.loc[data.Age.isnull() & (data.Title == 'Mr'), 'Age'] = age_by_title['Mr']
data.loc[data.Age.isnull() & (data.Title == 'Mrs'), 'Age'] = age_by_title['Mrs']
data.loc[data.Age.isnull() & (data.Title == 'Miss'), 'Age'] = age_by_title['Miss']
data.loc[data.Age.isnull() & (data.Title == 'Master'), 'Age'] = age_by_title['Master']
data.loc[data.Age.isnull() & (data.Title == 'Rare'), 'Age'] = age_by_title['Rare']
```

```python
data.isnull().sum()
```

```python
original[original.Embarked.isnull()]
```

```python
original.sort_values(by='Ticket').head(10)
```

Искам да заменя липсващите пристанища с най-логичната стойност - всички номера на билети 11ХХХХ са от 'S' или 'C', но тези около 113ХХХ, закупени от 'C' са на мъже, така, че липсващите стойности ще заменя с 'S' (или най-често срещаното пристанище, но реших да поразсъждавам).

```python
data.Embarked = data.Embarked.fillna('S')
```

```python
data.head(10)
```

Какво още трябва да направим:
- Да махнем <b>Name</b> и <b>Title</b>, тъй като извлякохме основното от тях
- Да енкоднем <b>Sex</b> и <b>Embarked</b> 

```python
data = data.drop(['Name', 'Title'],axis=1);
```

```python
data['Sex'] = data['Sex'].astype('category')
data['Sex'].cat.categories = [0,1]
data['Sex'] = data['Sex'].astype('int')
```

```python
data['EmbarkedC'] = (data.Embarked == 'C').astype(float)
data['EmbarkedQ'] = (data.Embarked == 'Q').astype(float)
data['EmbarkedS'] = (data.Embarked == 'S').astype(float)
data = data.drop('Embarked', axis=1)
```

```python
sns.countplot('Pclass', hue='Survived', data=data);
```

Виждаме, че умрелите са най-много от трета класа, което е и логично, тъй като трета класа е разположена най-ниско на кораба.
Ще енкоднем и класата.

```python
data['Class1'] = (data.Pclass == 1).astype(float)
data['Class2'] = (data.Pclass == 2).astype(float)
data['Class3'] = (data.Pclass == 3).astype(float)
data = data.drop('Pclass', axis=1)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), vmax=0.6, square=True, annot=True)
```

Уверяваме се, че хората от трета класа са имали най-малък шанс да оцелеят, а хората от първа - най-голям. Можем и да го визуализираме:

```python
data[['Survived','Age']].sort_values(by='Age').head(10)
```

```python
plt.rcParams['figure.figsize']=(50,50)
sns.countplot('Age', hue='Survived', data=data);
```

Ще разширя фийчърите за възраст, т.е. ще добавя няколко artificial. Ще опитам да разделя <b>Age</b> на два класа - деца (до 16г) и възрастни(над 16г), или на няколко интервала - \[0,16](повече оцеляват), \[16-32](повече умират),\[32-45],\[45-54],\[54-62](почти равно),\[62-80](умират).


```python
twoageranges = data.copy()
twoageranges.loc[twoageranges['Age'] <= 16, 'Age'] = 0
twoageranges.loc[twoageranges['Age'] > 16, 'Age'] = 1
twoageranges['isChild'] = (twoageranges.Age == 0).astype(float)
twoageranges['isAdult'] = (twoageranges.Age == 1).astype(float)
twoageranges.drop('Age', axis=1);
```

<b>Резултати за два класа за възрастта:</b>

```python
from sklearn.model_selection import train_test_split
X = twoageranges.drop('Survived', axis=1)
y = twoageranges['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=y)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)

print("train score:", model.score(X_train, y_train))
print("test score: ", model.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=0, n_estimators=70, max_depth=4).fit(X_train, y_train)
print("train score:", model.score(X_train, y_train))
print("test score: ", model.score(X_test, y_test))
```

```python
moreageranges = data.copy()
moreageranges.loc[moreageranges['Age'] <= 16, 'Age'] = 0
moreageranges.loc[(moreageranges['Age'] > 16) & (moreageranges['Age'] <=32), 'Age'] = 1
moreageranges.loc[(moreageranges['Age'] > 32) & (moreageranges['Age'] <=45), 'Age'] = 2
moreageranges.loc[(moreageranges['Age'] > 45) & (moreageranges['Age'] <=54), 'Age'] = 3
moreageranges.loc[(moreageranges['Age'] > 54) & (moreageranges['Age'] <=62), 'Age'] = 4
moreageranges.loc[moreageranges['Age'] > 62,'Age'] = 5
moreageranges['AgeRange1'] = (moreageranges.Age == 0).astype(float)
moreageranges['AgeRange2'] = (moreageranges.Age == 1).astype(float)
moreageranges['AgeRange3'] = (moreageranges.Age == 2).astype(float)
moreageranges['AgeRange4'] = (moreageranges.Age == 3).astype(float)
moreageranges['AgeRange5'] = (moreageranges.Age == 4).astype(float)
moreageranges['AgeRange6'] = (moreageranges.Age == 5).astype(float)
moreageranges.drop('Age',axis=1)
moreageranges.head(10)
```

<b>Резултати за много класове възраст:</b>

```python
X = moreageranges.drop('Survived', axis=1)
y = moreageranges['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=y)

model = LogisticRegression()
model.fit(X_train, y_train)

print("train score:", model.score(X_train, y_train))
print("test score: ", model.score(X_test, y_test))
```

```python
model = RandomForestClassifier(random_state=0, n_estimators=70, max_depth=4).fit(X_train, y_train)
print("train score:", model.score(X_train, y_train))
print("test score: ", model.score(X_test, y_test))
```

По-скоро разделянето на възрастта на множество интервали не ни е особено от полза, да разделена на три.

<b>Резултати от възрастта разделена на три интервала:</b>

```python
threeageranges = data.copy()
threeageranges.loc[threeageranges['Age'] <= 16, 'Age'] = 0
threeageranges.loc[(threeageranges['Age'] > 16) & (threeageranges['Age'] <=62), 'Age'] = 1
threeageranges.loc[moreageranges['Age'] > 62,'Age'] = 2
threeageranges['AgeRange1'] = (threeageranges.Age == 0).astype(float)
threeageranges['AgeRange2'] = (threeageranges.Age == 1).astype(float)
threeageranges['AgeRange3'] = (threeageranges.Age == 2).astype(float)
threeageranges = threeageranges.drop('Age',axis=1);
```

```python
X = threeageranges.drop('Survived', axis=1)
y = threeageranges['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=y)
model = RandomForestClassifier(n_estimators=70, max_depth=5).fit(X_train, y_train)
print("train score:", model.score(X_train, y_train))
print("test score: ", model.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
search = GridSearchCV(model, {'n_estimators': [10, 30, 50, 70, 100],
                              'max_depth': [3,4,5,6,7,8]})
search.fit(X, y)

pd.DataFrame(search.cv_results_)[['rank_test_score', 'mean_test_score', 'params']].sort_values(by='rank_test_score').head(10)
```

GridSearch връща леко различни (в хилядни) резултати при различните пускания, което е разбираемо, заради кросвалидацията. Но вкрайна сметка ще оставя <b>Age</b> разделена на три интервала.

Нека видим какво става ако добавим фийчъри дали пътникът е пътувал сам и ако не каква е големината на семейството му.
Пътинкът е пътувал сам когато сборът <b>Parch</b> (# of parents / children aboard) и <b>SibSp</b>(# of siblings / spouses aboard) равен на 0.

```python
newdata = threeageranges.copy()
newdata['FamilySize'] = newdata.Parch + newdata.SibSp + 1
newdata['IsAlone'] = (newdata.FamilySize == 0).astype(float)
```

```python
newdata = newdata.drop(['SibSp','Parch'],axis=1);
```

```python
X = newdata.drop('Survived', axis=1)
y = newdata['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=y)
model = RandomForestClassifier(n_estimators=70, max_depth=5).fit(X_train, y_train)
print("train score:", model.score(X_train, y_train))
print("test score: ", model.score(X_test, y_test))

search = GridSearchCV(model, {'n_estimators': [10, 30, 50, 70, 100],
                              'max_depth': [3,4,5,6,7,8]})
search.fit(X, y)

pd.DataFrame(search.cv_results_)[['rank_test_score', 'mean_test_score', 'params']].sort_values(by='rank_test_score').head(10)
```

След няколко пускания на gridseach-oвете, реших, че фийчърите за дали e пътувал сам човекът или със семейство, не ми подбряват резултата за модела, така че няма да ги включвам. 
Интересно, че пробвах с <b>SVC</b> и резултатите от gridseach-oвете за него бяха доста по-лоши ~ 0,75, което ми се стори странно, и ресърчвайки из нета, единственото обяснение, до което стигнах е, че дава лоши резултати, защото класовете не са балансирани.

Ще пробвам да вкарам още artificial фийчъри <b>Age\*Class</b> и <b>Fare\*Class</b>. За Fare*Class rational-a e, че би трябвало да има области на кораба от всяка класа, които са потънали първи и съответно повече хора са измряли от тях. 

```python
testdata = threeageranges.copy()
testdata['Fare*Class1'] = testdata['Fare']*testdata['Class1']
testdata['Fare*Class2'] = testdata['Fare']*testdata['Class2']
testdata['Fare*Class3'] = testdata['Fare']*testdata['Class3']
```

```python
X = testdata.drop('Survived', axis=1)
y = testdata['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=y)
model = RandomForestClassifier(n_estimators=70, max_depth=5).fit(X_train, y_train)

search = GridSearchCV(model, {'n_estimators': [10, 30, 50, 70, 100],
                              'max_depth': [3,4,5,6,7,8]})
search.fit(X, y)

pd.DataFrame(search.cv_results_)[['rank_test_score', 'mean_test_score', 'params']].sort_values(by='rank_test_score').head(5)
```
